```
#Ececução do download de imagens via multiplas threads
#processo sequencial
import threading #modulo para contrução de threads
import urllib.request #modulo para a requisição de url
import time

def downloadImagens(imagePath, fileName):
  print("Realizando o download.....", imagePath)
  urllib.request.urlretrieve(imagePath, fileName) #realiza a requisição paa a página da web
  print("Download Finalizado!")

def executeThread(i):
  imageName ='imagens_thread/image-'+ str(i) +".jpg"
  downloadImagens("http://lorempixel.com/400/200/sports", imageName)

t0= time.time() #armazena o tempo inicial de execução

threads = []#lista vazia que vai conter todas as threads criadas

#cria 10 threads, cada uma delas será responsável por realizar o download
for i in range(10):
  thread = threading.Thread(target=executeThread, args=(i,))
  threads.append(thread)
  thread.start()

#garante que as execuções foram finalizadas
for i in threads:
  i.join()

#calcula o tempo de execução
t1=time.time()
totalTime = t1 - t0
print("Tempo total de execução {}". format(totalTime))
```

**Exemplo paralelismo e concorrencia **

```
import threading #modulo para contrução de threads
import time
import random

counter = 10 #contador inicial

#função que adiciona um número para o contador

def tarefaA():
  global counter #variavel global
  while counter < 30:
    conuter +=1 
    print("A tarefaA incrementou o contador para {}".format(counter))
    sleepTime = random.randint(0,3) #escolhe randomicamente um valor entre 0 e 3 
    time.sleep(sleepTime)#coloca a thread para dormir

#função que retira um número do contador

def tarefaB():
  global counter #variavel global
  while counter > -30:
    counter -=1#decrementa o contador
    print("A tarefaB decrementou o contador para {}".format(counter))
    sleepTime = random.randint(0,3) #escolhe, randomicamente um valor entre 0 e 3
    time.sleep(sleepTime) #coloca o thread para dormir

t0 = time.time()
thread1 = threading.Thread(target=tarefaA) #intancia um obejto da classe Thread para executar a tarefa A

thread2 = threading.Thread(target=tarefaB) #intancia um obejto da classe Thread para executar a tarefa B


thread1.start() #inicia a thread1
thread2.start() #inicia a thread2

thread1.join() #certifica o fim da execução
thread2.join() #certifica o fim da execução

t1 = time.time()

print("Tempo total de execução {}".format(t1-t0))
```

**Estados de execução de uma Thread**

```
import threading
import time

#função que, simplesmente realiza o print de uma mensagem de execução
def threadWorker():
  #Neste ponto é onde ocorre a mudança do 'Runnable' para o 'Running'

  print("A thread entrou no estado 'Ruunning'")

  #quando chamamos a função time.sleep() a 
  #thread entra para o estado de not-running
  print("A thread entrou no estado 'Non-Running'")
  time.sleep(10)
  #quando a tarefa é finalizada a thread é terminada
  print("Execução da thread foi finalizada")

  #garbage collector

#neste momento a thread ainda não possui estados
#não existe alocação de recuros

print("Thread criada")
myThread = threading.Thread(target=threadWorker)

#quando é chamado o metodo myTread.start(), python realiza a 
#alocação de recursos e posteriormente passa para o estado de 
#"start" para o "Runnable" mas sem entrar em execução
print("Thread no estado 'Runnable'")
myThread.start()

#quando o metodo join é chamado, a thread passa para o estado
#"terminated"
myThread.join()
print("a thread está no estado de 'Terminated")
```

**Em qual Thread estamos ?**

```
import threading
import time

def threadTarget():
  print("Thread atual: {}".format(threading.current_thread()))

threads = []

for i in range(10):
  thread = threading.Thread(target=threadTarget)
  thread.start()
  threads.append(thread)

for thread in threads:
  thread.join()
```

**Main Thread**

```
import threading
import time

def myChildThread():
  print("Thread filha inicializada.....")
  time.sleep(5)
  print("Thread atual....")
  print(threading.current_thread())
  print("-------------------------")
  print("Main Thread -----")
  print(threading.main_thread())
  print("-----------------------")
  print("Thread filha finalizada")


child = threading.Thread(target=myChildThread)
child.start()
child.join()
```
