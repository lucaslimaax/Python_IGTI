```
import pandas as pd

#leitura dos dados csv

df = pd.read_csv("https://pycourse.s3.amazonaws.com/temperature.csv")
df

df.sort_values?
```

```
#plot de linhas
df.plot();

```

```
#plot de linhas: tamanho
#colocar ; ao final para esconder as informações adicionais

df.plot(figsize=(10, 5));
```

```
#plot de linas: grid

df.plot(figsize=(10, 5), grid=True);
```

```
#plot de linhas: styles

# df.plot(style='-o', figsize=(10, 5), grid=True);

df.plot(style='--', figsize=(10, 5), grid=True);
```

```
#plot de linhas: linewidth

df.plot(style='-o', linewidth=2.5, figsize=(10, 5), grid=True)

```

```
#plot de linhas: color 

df.plot(style='-o', linewidth=2.5, figsize=(10, 5), color='red', grid=True)
```

```
#plot cor da linha: hexadecimal

df.plot(style='-o', linewidth=2.5, figsize=(10, 5), color='#54f542', grid=True)
```

```
#plot de barras

df['classification'].value_counts().plot.bar(figsize=(10, 5), rot=0);
```

```
#plot de pizza
#autopct='%1.1f%%'   shadow= True = colocar as porcentagens com uma casa decimal

df['classification'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(10, 7));
```
