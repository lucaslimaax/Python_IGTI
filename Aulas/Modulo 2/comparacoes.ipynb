```
  #criação dos arrays
  import numpy as np

  x = np.array([[1, 2], [3, 4]])
  y = np.array([1.5, 3.5])
  print("x: \n", x)
  print("y: \n", y)

  
```

```
#comparações ponto a ponto

print("Comparação de um array com um escalar (>): \n", x > 2)
print("Comparação de um array com um escalar (>=): \n", x >=2)
print("Comparação de um array com um escalar (<): \n", x < 2)
print("Comparação de um array com um escalar (<=): \n", x <= 2)
print("Comparação entre arrays (==): \n", x == x)
print("Comparação entre arrays (>): \n", x == x)
print("Comparação entre arrays (>): \n", x > y) #broadcasting

```

```
#indexação boleana

x = np.array([[1, 3, 7],
              [4, 11, 21],
              [42, 8, 9]])
print("x:\n", x) 
```

```
#indexão booleana: retornar o número de elementos maiores que k

k = 10
cond = x > k
print("condição: \n", cond)
print(f"elementos maiores que {k}:", x[cond])
# print(f"número de emlementos maiores que {k}:", len(x[cond]))
```
