```
#criação de dois arrays x e y

import numpy as np

x=np.ones((2,2))
y=np.eye(2)
print("x: \n", x)
print("y: \n", y)

```

```
#soma 
print("soma de dois arrays: \n", x+y)
print("soma com float/int:\n", x+2.) #broadcasting(possivel somar com outros tipos de elementos)
```

```
#subtração
print("subtração de dois arrays:\n", x-y)
print("subtração com float/int:\n", x - 2) #broadcasting
```

```
#divisão

print("divisão de dois arrays:\n", x/y)
print("divisão com float/int:\n", x/2)
```

```
#soma, subtração e divisão

print("Combinação de operadores: \n", (x+y)/(x-2))
```

```
#Multiplicação

print("multiplicação de dois arrays:\n", x * y)
print("multiplicação com float/int:\n", x * 2) #broadcasting
```

```
#criar arrays costumizados

5 * np.ones((10, 10))
```

```
#definição do problema

A = np.array([[1, 2], [3, -2]])
c = np.array([[7], [-11]])
print("A: \n", A)
print("c: \n", c)
```

```
#Sistemas de equações

#solução

#np.linalg é um conjunto de metodos para trabalhar com algebra linear
# x = np.dot(np.linalg.inv(A), c)
x = np.linalg.inv(A) @ c #solução usando o operador @ do próprio python q multiplica matrizes
print("(a, b):", x.ravel())
```
